```python
import pandas as pd
```

```python
with open('The-Secret-in-Their-Eyes_2009.txt', 'r') as fd:
    all_the_lines = fd.readlines()
```

```python
len(all_the_lines)
```

```python
from dateutil import parser
from datetime import timedelta
import datetime

delta = -10

all_the_new_lines = []

for l in all_the_lines:
    if ' --> ' in l:
        b,e = l.split(' --> ')
        bb = parser.parse(b)
        ee = parser.parse(e)
        bbb = bb + datetime.timedelta(seconds=delta)
        eee = ee + datetime.timedelta(seconds=delta)
        bbbb = bbb.strftime("%H:%M:%S,%f")[:12]
        eeee = eee.strftime("%H:%M:%S,%f")[:12]
        
        ll = '{} --> {}\n'.format(bbbb,eeee)
        
        print('Old:', l.rstrip())
        print('New:', ll.rstrip())
        
        all_the_new_lines.append(ll)
    else:
        all_the_new_lines.append(l)
```

```python
with open('The-Secret-in-Their-Eyes_2009_Meli2.txt', 'w') as fd:
    fd.write(''.join(all_the_new_lines))
```
